```python
import matplotlib.pyplot as plt
import random
```

## One Dimension

```python
random.seed(123)
x = random.sample(range(1, 101), 50)
print(x)
```

```python
y = [0 for var in x]
print(y)
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(12, 1)
plt.scatter(x, y)

for grid_pt in [20, 40, 60, 80]:
    plt.axvline(x=grid_pt, color="silver")
    
ax.set_xlabel("Dim #1", fontsize=12)
plt.yticks([], [])
plt.title("1D", fontsize=15)
plt.show()
```

## Two Dimension

## Three Dimension

```python
random.seed(123)
x = random.sample(range(1, 101), 50)
y = random.sample(range(1, 101), 50)
z = random.sample(range(1, 101), 50)

print("x >>> ", x)
print("y >>> ", y)
print("z >>> ", z)
```

```python
from mpl_toolkits import mplot3d
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1, projection = '3d')

fig.set_size_inches(8,8)
ax.scatter(x, y, z)
    
ax.set_xlabel("Dim #1", fontsize=12)
ax.set_ylabel("Dim #2", fontsize=12)
ax.set_zlabel("Dim #3", fontsize=12)

plt.title("2D", fontsize=15)
plt.show()
```
